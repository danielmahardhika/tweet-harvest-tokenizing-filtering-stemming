```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Ambil stopwords bahasa Indonesia dengan Pysastrawi
stopword_factory = StopWordRemoverFactory()
stopwords = stopword_factory.get_stop_words()

# Gunakan Pysastrawi untuk stemming
stemmer_factory = StemmerFactory()
stemmer = stemmer_factory.create_stemmer()
# Gunakan WordNetLemmatizer untuk lemmatisasi
lemmatizer = nltk.stem.WordNetLemmatizer()

custom_stopwords = [
    # Singkatan umum
    'yg', 'dgn', 'gt', 'gtu', 'gitu', 'jg', 'juga', 'dr', 'dari', 'utk', 'untuk',
    'dlm', 'dalam', 'pd', 'pada', 'krn', 'karena', 'spy', 'supaya', 'trs', 'terus',
    'aja', 'sih', 'nya', 'ni', 'deh', 'sih', 'lho', 'lah', 'kalo', 'klo', 'gak', 'ga',
    'ngga', 'enggak', 'gk', 'yah', 'ya', 'si', 'dg', 'ny', 'lg', 'lagi', 'udh', 'sdh',
    'udah', 'sudah', 'nih', 'tuh', 'bs', 'bisa', 'kyk', 'kyak', 'kayak', 'spt', 'seperti',
    'bgtu', 'begitu', 'bgt', 'banget', 'org', 'orang', 'jd', 'jadi', 'tp', 'tapi', 'cm', 'cuma',
    'loo', 'yuk', 'yu', 'dak', 'wes', 'ampe', 'amet', 'amit', 'yaa', 'tpi', 'euy', 'hayu',

    # Kata-kata yang sering muncul
    'bang', 'mas', 'mbak', 'kakak', 'adik', 'bro', 'sis', 'pak', 'bu', 'om', 'tante',
    'ayah', 'ibu', 'mama', 'papa', 'nenek', 'kakek', 'teteh', 'aa', 'abang', 'emak',
    'rek', 'cuy', 'sob', 'gan', 'agan', 'kang', 'mba', 'neng', 'ning', 'den', 'gus',

    # Kata sapaan dan istilah media sosial
    'min', 'admin', 'guys', 'gaes', 'bro', 'sis', 'follower', 'followers', 'follow',
    'dm', 'mention', 'share', 'like', 'komen', 'comment', 'posting', 'post', 'timeline',
    'story', 'ig', 'fb', 'twitter', 'tiktok', 'youtube', 'yt', 'video', 'foto', 'pict',

    # Kata yang berkaitan dengan waktu
    'kemarin', 'besok', 'lusa', 'sekarang', 'tadi', 'nanti', 'pagi', 'siang', 'sore', 'malam',
    'senin', 'selasa', 'rabu', 'kamis', 'jumat', 'sabtu', 'minggu', 'januari', 'februari', 'maret',
    'april', 'mei', 'juni', 'juli', 'agustus', 'september', 'oktober', 'november', 'desember',

    # Kata perintah
    'tolong', 'bantu', 'mohon', 'harap', 'coba', 'silakan', 'sila', 'mari', 'ayo', 'yuk',

    # Kata penghubung
    'dan', 'atau', 'tetapi', 'namun', 'serta', 'dengan', 'padahal', 'sedangkan', 'sambil',
    'ketika', 'selama', 'setelah', 'sebelum', 'sejak', 'hingga', 'sampai', 'sehingga', 'karena',
    'sebab', 'oleh', 'untuk', 'bagi', 'demi', 'tentang', 'mengenai', 'terhadap', 'pada',

    # Kata tanya
    'apa', 'siapa', 'kapan', 'dimana', 'kemana', 'darimana', 'bagaimana', 'mengapa', 'kenapa',
    'gimana', 'dmn', 'kmn', 'dmn', 'brp', 'berapa',

    # Interjeksi dan kata seru
    'wah', 'aduh', 'astaga', 'waduh', 'duh', 'wow', 'eh', 'oh', 'hah', 'hmm', 'hm', 'hmm',
    'loh', 'lho', 'lha', 'hei', 'hai', 'halo', 'hello', 'hi', 'hola', 'oy', 'woy', 'coy',

    # Kata-kata umum yang sering muncul tapi tidak penting
    'adalah', 'ialah', 'yaitu', 'yakni', 'merupakan', 'itu', 'ini', 'dia', 'beliau', 'mereka',
    'kita', 'kami', 'saya', 'aku', 'gue', 'gw', 'lu', 'lo', 'kamu', 'kalian', 'dia', 'beliau',
    'anda', 'bapak', 'ibu', 'pak', 'bu', 'mas', 'mbak', 'om', 'tante', 'kakak', 'adik',

    # Kata-kata spesifik dalam dataset
    'mbg', 'http', 'tco', 'amp', 'tempo', 'tempodotco', 'bpk', 'icw', 'bumn', 'bgn', 'indonesiaemas',
    'prabowo', 'gibran', 'guardian', 'bbc', 'malay', 'subianto', 'awesomeintelligence', 'vietnam', 'mail',
    'south', 'china', 'morning', 'afternoon', 'evening', 'night', 'monday', 'tuesday', 'wednesday',
    'thursday', 'friday', 'saturday', 'sunday', 'jilmekin', 'jilmek', 'late', 'taklukkanutbk', 'vcsbugilll',
    'unk', 'vcscrottt', 'vcs', 'soft', 'stunting', 'pic', 'penuhigiziindonesia', 'tempoplus'
]

# Gabungkan stopwords yang sudah ada dengan custom_stopwords
stopwords.extend(custom_stopwords)
stopwords = list(set(stopwords))  # Hapus duplikat

# Dictionary singkatan Indonesia yang lebih lengkap
abbreviation_dict = {
    'mbg': 'makan bergizi gratis',
    'mbeg': 'makan bergizi gratis',
    'mbtg': 'makan bergizi gratis',
    'icw': 'indonesia corruption watch',
    'bgn': 'badan gizi nasional',
    'bpk': 'badan pemeriksa keuangan',
    'klb': 'kejadian luar biasa',
    'sd': 'sekolah dasar',
    'smp': 'sekolah menengah pertama',
    'sma': 'sekolah menengah atas',
    'skh': 'sekolah khusus',
    'man': 'madrasah aliyah negeri',
    'yg': 'yang',
    'gt': 'gitu',
    'dll': 'dan lain lain',
    'dsb': 'dan sebagainya',
    'ruu': 'rancangan undang undang',
    'dgn': 'dengan',
    'dg': 'dengan',
    'utk': 'untuk',
    'dr': 'dari',
    'pd': 'pada',
    'krn': 'karena',
    'jg': 'juga',
    'tdk': 'tidak',
    'org': 'orang',
    'bs': 'bisa',
    'jd': 'jadi',
    'tp': 'tapi',
    'tpi': 'tapi',
    'ttg': 'tentang',
    'scr': 'secara',
    'dpt': 'dapat',
    'sdh': 'sudah',
    'udh': 'sudah',
    'trs': 'terus',
    'spy': 'supaya',
    'hrs': 'harus',
    'bgtu': 'begitu',
    'bgt': 'banget',
    'sblm': 'sebelum',
    'stlh': 'setelah',
    'jk': 'jika',
    'msk': 'masuk',
    'blm': 'belum',
    'kmrn': 'kemarin',
    'skrg': 'sekarang',
    'tsb': 'tersebut',
    'tgl': 'tanggal',
    'bln': 'bulan',
    'thn': 'tahun',
    'mgg': 'minggu',
    'smp': 'sampai',
    'brp': 'berapa',
    'cb': 'coba',
    'byk': 'banyak',
    'skt': 'sakit',
    'pkl': 'pukul',
    'cm': 'cuma',
    'sy': 'saya',
    'gw': 'saya',
    'gue': 'saya',
    'aku': 'saya',
    'lo': 'kamu',
    'lu': 'kamu',
    'u': 'kamu',
    'kk': 'kakak',
    'dek': 'adik',
    'no': 'nomor',
    'jwt': 'jawat',
    'tmpt': 'tempat',
    'tlp': 'telepon',
    'tlpn': 'telepon',
    'hp': 'handphone',
    'wa': 'whatsapp',
    'fb': 'facebook',
    'ig': 'instagram',
    'ln': 'luar negeri',
    'rb': 'ribu',
    'jt': 'juta',
    'M': 'juta',
    'T': 'triliun',
    'rp': 'rupiah',
    'kec': 'kecamatan',
    'kab': 'kabupaten',
    'prov': 'provinsi',
    'rt': 'rukun tetangga',
    'rw': 'rukun warga',
    'dpr': 'dewan perwakilan rakyat',
    'pemda': 'pemerintah daerah',
    'pemkot': 'pemerintah kota',
    'pemkab': 'pemerintah kabupaten',
    'na': 'national',
    'pd': 'partai demokrat',
    'coy': 'kawan',
    'sob': 'sobat',
    'dmn': 'dimana',
    'kmn': 'kemana',
    'gmn': 'bagaimana',
    'kpn': 'kapan',
    'ampe': 'sampai',
    'rame': 'ramai',
    'gk': 'tidak',
    'ga': 'tidak',
    'gak': 'tidak',
    'ngga': 'tidak',
    'kalo': 'kalau',
    'kl': 'kalau',
    'klo': 'kalau',
    'nih': 'ini',
    'ni': 'ini',
    'tuh': 'itu',
    'tu': 'itu',
    'dah': 'sudah',
    'udah': 'sudah',
    'dg': 'dengan',
    'ama': 'sama',
    'bareng': 'bersama',
    'bbrp': 'beberapa',
    'sbnrnya': 'sebenarnya',
    'stlh': 'setelah',
    'sblm': 'sebelum',
    'skrng': 'sekarang',
    'lgsg': 'langsung',
    'spt': 'seperti',
    'kyak': 'seperti',
    'kyk': 'seperti',
    'trm': 'terima',
    'makasih': 'terima kasih',
    'tq': 'terima kasih',
    'thx': 'terima kasih',
    'hayu': 'ayo',
    'gas': 'ayo',
    'yuk': 'ayo',
    'yu': 'ayo',
    'kmrin': 'kemarin',
    'dpat': 'dapat',
    'mlm': 'malam',
}

# Daftar reguler expression patterns untuk membersihkan teks
patterns = [
    # URLs
    (r'https?://\S+', ''),
    (r'www\.\S+', ''),
    # HTTP tco links (Twitter)
    (r'http\s*-?\s*tco\w+', ''),
    # Email
    (r'\S+@\S+\.\S+', ''),
    # Hapus semua http dan tco
    (r'\bhttp\b', ''),
    (r'\btco\b', ''),
    # Hapus angka
    (r'\d+', ''),
    # Hapus karakter khusus dan simbol
    (r'[^\w\s]', ' '),
    # Hapus spasi berlebih
    (r'\s+', ' '),
    # Hapus anomali tc diikuti huruf lainnya (seperti tcovxftyxgp, tcoitjcmjo)
    (r'\btco[a-z]+\b', ''),
    # Hapus anomali lainnya
    (r'\b[a-z]{1,2}\b', ''),  # Kata dengan 1-2 huruf
    (r'\b[a-z]{15,}\b', ''),  # Kata dengan 15+ huruf
    (r'\b([a-z])\1{2,}\b', ''),  # Karakter berulang 3+ kali
]

# Fungsi untuk menghapus aksen
def remove_accents(data):
    return ''.join(x for x in unicodedata.normalize('NFKD', data) if x in string.ascii_letters or x == " ")

# Fungsi untuk menerapkan pola regex cleaning
def apply_regex_cleaning(text):
    for pattern, replacement in patterns:
        text = re.sub(pattern, replacement, text)
    return text.strip()

# Fungsi untuk mengubah singkatan menjadi bentuk lengkap
def expand_abbreviations(text):
    words = text.split()
    expanded_words = []

    for word in words:
        if word.lower() in abbreviation_dict:
            expanded_words.append(abbreviation_dict[word.lower()])
        else:
            expanded_words.append(word)

    return ' '.join(expanded_words)

# Fungsi untuk menghilangkan kata-kata duplikasi berurutan
def remove_consecutive_duplicates(text):
    words = text.split()
    result = []
    prev_word = None

    for word in words:
        if word != prev_word:
            result.append(word)
        prev_word = word

    return ' '.join(result)

# Fungsi untuk menghilangkan duplikasi kata dalam seluruh kalimat
def remove_all_duplicates(text):
    words = text.split()
    unique_words = []
    seen = set()

    for word in words:
        if word not in seen:
            unique_words.append(word)
            seen.add(word)

    return ' '.join(unique_words)

# Regular expression untuk mencocokkan kata yang valid
RE_VALID = "^[^" + string.punctuation + "]+$"
# Minimal panjang string yang diperbolehkan
MIN_STR_LEN = 3
# Daftar tipe POS (Part Of Speech) yang dikenali oleh WordNet
POS_TYPES = ['NN', 'JJ', 'VB', 'RB']
# Daftar tipe POS yang akan diubah menjadi 'n' untuk digunakan oleh WordNetLemmatizer
DI_POS_TYPES = {'N': 'n', 'J': 'a', 'V': 'v', 'R': 'r'}

# Proses semua quotes
li_tokens = []
li_token_lists = []
li_lem_strings = []

for text in li_quotes:
    # Bersihkan teks dengan regex patterns
    text = apply_regex_cleaning(text)

    # Ubah singkatan menjadi bentuk lengkap
    text = expand_abbreviations(text)

    # Hilangkan kata duplikasi berurutan
    text = remove_consecutive_duplicates(text)

    # Opsional: Hilangkan semua kata duplikasi dalam seluruh teks
    text = remove_all_duplicates(text)

    # Tokenisasi per kalimat, kemudian jadikan lowercase
    tokens = [word.lower() for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]

    # Proses semua token per quote
    li_tokens_quote = []
    li_tokens_quote_lem = []
    for token in tokens:
        # Hapus aksesn
        t = remove_accents(token)

        # Hapus punctuation/tanda baca
        t = str(t).translate(string.punctuation)
        li_tokens_quote.append(t)

        # Tambahkan token yang mewakili "tidak ada kecocokan lemmatisasi"
        li_tokens_quote_lem.append("-")  # token ini akan dihapus jika ditemukan kecocokan lemmatisasi di bawah

        # Proses setiap token
        if t not in stopwords:
            if re.search(RE_VALID, t):
                if len(t) >= MIN_STR_LEN:
                    # Ingat bahwa POS (Part Of Speech) diperlukan sebagai masukan ke lemmatizer
                    # (jika tidak maka diasumsikan kata tersebut adalah kata benda)
                    pos = nltk.pos_tag([t])[0][1][:2]
                    pos2 = 'n'  # set default ke noun atau kata benda
                    if pos in DI_POS_TYPES:
                        pos2 = DI_POS_TYPES[pos]

                    stem = stemmer.stem(t)
                    lem = lemmatizer.lemmatize(t, pos=pos2)  # lemmatize dengan POS yang benar

                    if pos in POS_TYPES:
                        li_tokens.append((t, stem, lem, pos))

                        # Hapus token "-" token dan tambahkan kecocokan lemmatisasi
                        li_tokens_quote_lem = li_tokens_quote_lem[:-1]
                        li_tokens_quote_lem.append(lem)

    # Buat daftar daftar token dari token yang dilemmatisasi
    li_token_lists.append(li_tokens_quote)

    # Buat daftar string dari token yang dilemmatisasi
    str_li_tokens_quote_lem = ' '.join(li_tokens_quote_lem)
    li_lem_strings.append(str_li_tokens_quote_lem)

# Bikin dataframes dari lists
df_token_lists = pd.DataFrame(li_token_lists)

print("Tokenisasi:")
print(df_token_lists.head(10).to_string())

# Ganti "None" dengan string kosong
for c in df_token_lists:
    if str(df_token_lists[c].dtype) in ('object', 'string_', 'unicode_'):
        df_token_lists[c].fillna(value='', inplace=True)

df_lem_strings = pd.DataFrame(li_lem_strings, columns=['lem quote'])

print("\n\n")
print("Cleaning:")
print(df_lem_strings.head(10).to_string())
```
